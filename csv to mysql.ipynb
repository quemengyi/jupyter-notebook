```python
import pymysql
import os
import pandas as pd

try:
    conn = pymysql.connect(host='localhost', user='root', password='root', db='csv', charset='utf8')
    cur = conn.cursor()
    print('数据库连接成功！')
    print(' ')
except:
    print('数据库连接失败！')

path = os.getcwd()
files = os.listdir(path)
for file in files:
  if file.split('.')[-1] in ['csv']:
    filename = file.split('.')[0]
    filename = 'tab_' + filename
    
    f = pd.read_csv(file, encoding='gbk')
    columns = f.columns.tolist()
    types = f.dtypes
    field = []
    table = []
    for item in columns:
        if 'int' in str(types[item]):
            char = item + ' INT'
        elif 'float' in str(types[item]):
            char = item + ' DOUBLE'
        elif 'object' in str(types[item]):
            char = item + ' VARCHAR(255)'
        elif 'datetime' in str(types[item]):
            char = item + ' DATETIME'
        else:
            char = item + ' VARCHAR(255)'
        field.append(item)
        table.append(char)
    fields = ','.join(field)
    tables = ','.join(table)
    
    table_sql = 'CREATE TABLE IF NOT EXISTS ' + filename + '(' + 'id0 int PRIMARY KEY NOT NULL auto_increment,' + tables + ');'
    print('表:' + filename + ',开始创建......')
    cur.execute(table_sql)
    conn.commit()
    print('表:' + filename + ',创建成功！')
    
    print('表:' + filename + ',开始插入数据......')
    s = ','.join(['%s' for _ in range(len(f.columns))])
    insert_sql = 'insert into {}({}) values({})'.format(filename,fields,s)
    values = f.values.tolist()
    cur.executemany(insert_sql, values)
    conn.commit()
    print('表:' + filename + ',插入数据完成！')
    print(' ')

cur.close()
conn.close()
```
